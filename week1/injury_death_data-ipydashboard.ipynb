```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import numpy as np
from ipywidgets import interact

xl = pd.ExcelFile('week1.xlsx')
xl.sheet_names
```

```python
def get_intent(intent):
    if intent == 'Homicide':
        return 1
    elif intent == 'Legal intervention/war':
        return 2
    elif intent == 'Suicide':
        return 3
    elif intent == 'Undetermined':
        return 4
    else:
        return 5

def get_age(age):
    if age == '< 15':
        return 15
    elif age == '15â€“24':
        return 24
    elif age == '25â€“44':
        return 44
    elif age == '45â€“64':
        return 64
    elif age == '65â€“74':
        return 74
    else:
        return 75

df = xl.parse('NCHS_-_Injury_Mortality__United')
df.head()
```

```python
@interact(hue=['Injury Mechanism', 'Injury Intent', 'Age Group (Years)', 'Sex', 'Gender-Age', 'Gender-Intent'])

def func(hue):
    if hue == 'Injury Mechanism':
        row = df['Injury Mechanism'] != 'All Mechanisms'
        df_temp = df[row]
        df_IM = df_temp.groupby(['Injury Mechanism']).agg([np.sum])['Deaths']
        df_IM.plot(kind='pie', y='sum', autopct='%1.1f%%', title='Deaths vs. Injury Mechanism', ylabel='', labeldistance=None, figsize=(12, 8)).legend(bbox_to_anchor=(1, 1.02), loc='upper left')
    elif hue == 'Injury Intent':
        row = df['Injury Intent'] != 'All Intentions'
        df_temp = df[row]
        df_II = df_temp.groupby(['Injury Intent']).agg([np.sum])['Deaths']
        df_II.plot(kind='pie', y='sum', autopct='%1.1f%%', title='Deaths vs. Injury Intent', ylabel='', labeldistance=None, figsize=(12, 8)).legend(bbox_to_anchor=(1, 1.02), loc='upper left')
    elif hue == 'Age Group (Years)':
        row = df['Age Group (Years)'] != 'All Ages'
        df_temp = df[row]
        df_ag = df_temp.groupby(['Age Group (Years)']).agg([np.sum])['Deaths']
        df_ag.plot(kind='pie', y='sum', autopct='%1.1f%%', title='Deaths vs. Age Group', ylabel='', labeldistance=None, figsize=(12, 8)).legend(bbox_to_anchor=(1, 1.02), loc='upper left')
    elif hue == 'Sex':
        row = df['Sex'] != 'Both sexes'
        df_temp = df[row]
        df_sex = df_temp.groupby(['Sex']).agg([np.sum])['Deaths']
        df_sex.plot(kind='pie', y='sum', autopct='%1.1f%%', title='Deaths vs. Sex', ylabel='', labeldistance=None, figsize=(12, 8)).legend(bbox_to_anchor=(1, 1.02), loc='upper left')
    elif hue == 'Gender-Age':
        df_temp = df[df['Age Group (Years)'] != 'All Ages']
        df_temp = df_temp[df_temp['Sex'] != 'Both sexes']
        df_temp['Age'] = df_temp['Age Group (Years)'].apply(get_age)
        df_temp = pd.concat([df_temp['Deaths'], df_temp['Sex'], df_temp['Age']], axis = 1)
        df_temp = df_temp.groupby(['Age', 'Sex']).agg([np.sum])['Deaths'].reset_index()
        df_temp.rename(columns = {"sum":"Deaths"}, inplace = True)

        sns.pairplot(df_temp, hue = 'Sex')
        plt.show()
    else:
        df_temp = df[df['Injury Intent'] != 'All Intentions']
        df_temp = df_temp[df_temp['Sex'] != 'Both sexes']
        df_temp['Intent'] = df_temp['Injury Intent'].apply(get_intent)
        df_temp = pd.concat([df_temp['Deaths'], df_temp['Sex'], df_temp['Intent']], axis = 1)
        df_temp = df_temp.groupby(['Intent', 'Sex']).agg([np.sum])['Deaths'].reset_index()
        df_temp.rename(columns = {"sum":"Deaths"}, inplace = True)

        sns.pairplot(df_temp, hue = 'Sex')
        plt.show()
```

Run all cells to generate data. Change hue in the last block to view different data plots. 
Note that for Gender-Intent, the different injury intents are represented by below orders in order to generate plots. 
    1 - Homicide
    2 - Legal intervention/war
    3 - Suicide
    4 - Undetermined
    5 - Unintentional
