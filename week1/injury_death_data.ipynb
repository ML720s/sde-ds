```python
import pandas as pd
import matplotlib as plt

xl = pd.ExcelFile('week1.xlsx')
xl.sheet_names
```

```python
df = xl.parse('NCHS_-_Injury_Mortality__United')
df.head()
```

```python
print(df.dtypes)
```

```python
import numpy as np

row = df['Injury Mechanism'] != 'All Mechanisms'
df_temp = df[row]
df_IM = df_temp.groupby(['Injury Mechanism']).agg([np.sum])['Deaths']
df_IM.plot(kind='pie', y='sum', autopct='%1.1f%%', title='Deaths vs. Injury Mechanism', ylabel='', labeldistance=None, figsize=(12, 8)).legend(bbox_to_anchor=(1, 1.02), loc='upper left')
```

```python
row = df['Injury Intent'] != 'All Intentions'
df_temp = df[row]
df_II = df_temp.groupby(['Injury Intent']).agg([np.sum])['Deaths']
df_II.plot(kind='pie', y='sum', autopct='%1.1f%%', title='Deaths vs. Injury Intent', ylabel='', labeldistance=None, figsize=(12, 8)).legend(bbox_to_anchor=(1, 1.02), loc='upper left')
```

```python
row = df['Age Group (Years)'] != 'All Ages'
df_temp = df[row]
df_ag = df_temp.groupby(['Age Group (Years)']).agg([np.sum])['Deaths']
df_ag.plot(kind='pie', y='sum', autopct='%1.1f%%', title='Deaths vs. Age Group', ylabel='', labeldistance=None, figsize=(12, 8)).legend(bbox_to_anchor=(1, 1.02), loc='upper left')
```
