```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("task3.csv")
```

```python
df1 = df.copy()
df1.loc['Sum'] = df1.sum()
df_temp = df1.loc['Sum']['Anti-Female':'Anti-Mental Disability'].reset_index().sort_values(by='Sum', ascending=False)
```

```python
df_content = df_temp.head(10)
df_content.plot(kind='pie', y='Sum', autopct='%1.1f%%', title='Top hates', ylabel='', labeldistance=None, figsize=(9, 6))
plt.legend(df_content['index'], bbox_to_anchor=(1, 1.02), loc='upper left')
```

```python
df2 = df.groupby(['County', 'Year']).agg([np.sum])['Total Offenders'].reset_index()
sns.pairplot(df2, hue = 'County', height = 10, diag_kind = 'hist')
plt.show()
```
